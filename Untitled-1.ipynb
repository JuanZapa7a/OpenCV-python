```python
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage.io import imread, imshow
from skimage.segmentation import chan_vese

fig, axes = plt.subplots(1, 3, figsize=(10, 10))
 
# Sample Image of scikit-image package
astronaut = imread('images/foto11.jpg')
gray_astronaut = rgb2gray(astronaut)

iterations=1000
 
# Computing the Chan VESE segmentation technique
chanvese_gray_astronaut = chan_vese(gray_astronaut,
                                    max_iter=iterations,
                                    extended_output=True)
 
ax = axes.flatten()
 
# Plotting the original image
ax[0].imshow(gray_astronaut, cmap="gray")
ax[0].set_title("Original Image")
 
# Plotting the segmented - 100 iterations image
ax[1].imshow(chanvese_gray_astronaut[0], cmap="gray")
title = "Chan-Vese segmentation - {} iterations".format(iterations)
format(len(chanvese_gray_astronaut[2]))
 
ax[1].set_title(title)
 
# Plotting the final level set
ax[2].imshow(chanvese_gray_astronaut[1], cmap="gray")
ax[2].set_title("Final Level Set")
plt.show()
```

```python
# Importing required libraries
from skimage.segmentation import slic, mark_boundaries
from skimage.data import astronaut
from skimage.color import rgb2hsv, rgb2gray, label2rgb
from skimage import data, img_as_float

# Setting the plot size as 15, 15
plt.figure(figsize=(15,15))
 
# Sample Image of scikit-image package
astronaut = img_as_float(imread('images/foto11.jpg'))
 
# Applying Simple Linear Iterative
# Clustering on the image
# - 50 segments & compactness = 10
astronaut_segments = slic(astronaut,
                          n_segments=1000,
                          compactness=10,
                          start_label=1,sigma=1, enforce_connectivity=True, convert2lab=True)
plt.subplot(1,3,1)
 
# Plotting the original image
plt.imshow(astronaut)
plt.subplot(1,3,2)
 
# Converts a label image into
# an RGB color image for visualizing
# the labeled regions.
image = label2rgb(astronaut_segments,
                     astronaut,
                     kind = 'avg', bg_label=1)
plt.imshow(image)

plt.subplot(1,3,3)

plt.imshow(mark_boundaries(astronaut, astronaut_segments, mode='thick'))
```

```python
image_hsv = rgb2hsv(image)

fig, ax = plt.subplots(1, 3, figsize=(12,4))
ax[0].imshow(image_hsv[:,:,0], cmap='gray')
ax[0].set_title('Hue')
ax[1].imshow(image_hsv[:,:,1], cmap='gray')
ax[1].set_title('Saturation')
ax[2].imshow(image_hsv[:,:,2], cmap='gray')
ax[2].set_title('Value');
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(image_hsv[:,:,0],cmap='hsv')
ax[0].set_title('hue')

ax[1].imshow(image_hsv[:,:,1],cmap='hsv')
ax[1].set_title('transparency')

ax[2].imshow(image_hsv[:,:,2],cmap='hsv')
ax[2].set_title('value')

fig.colorbar(imshow(image_hsv[:,:,0],cmap='hsv')) 
fig.tight_layout()
```

```python
import numpy as np

#refer to hue channel (in the colorbar)
lower_mask = image_hsv[:,:,0] > 0.1

#refer to hue channel (in the colorbar)
upper_mask = image_hsv[:,:,0] < 0.18

#refer to transparency channel (in the colorbar)
saturation_mask = image_hsv[:,:,1] > 0.57
 
mask = upper_mask*lower_mask*saturation_mask

red = image[:,:,0]*mask
green = image[:,:,1]*mask
blue = image[:,:,2]*mask
image_masked = np.dstack((red,green,blue))
imshow(image_masked)
```

```python
from skimage import filters


# Setting plot size to 15, 15
plt.figure(figsize=(15, 15))

# To gray
gray_image_masked = rgb2gray(image_masked)

# Computing Otsu's thresholding value
threshold = filters.threshold_otsu(gray_image_masked)

# Computing binarized values using the obtained
# thresold
image_bw = (gray_image_masked > threshold)*1
plt.subplot(2,2,1)
plt.title("Threshold: >"+str(threshold))
# Displaying the binarized image
plt.imshow(image_bw, cmap = "gray")
```

```python
from skimage import morphology

def multi_dil(im,num):
  for i in range(num):
      im = morphology.binary_dilation(im)
  return im

def multi_ero(im,num):
  for i in range(num):
      im = morphology.binary_erosion(im)
  return im
```

```python
image_cleaned = multi_ero(multi_dil(image_bw,20),20)
plt.imshow(image_cleaned,cmap='gray')
```

```python
from skimage.segmentation import clear_border

# remove artifacts connected to image border
image_cleaned = clear_border(image_cleaned)
plt.imshow(image_cleaned,cmap='gray')
```

```python
from skimage.measure import label, regionprops

label_im = label(image_cleaned)
imshow(label_im*255.0)
```

```python
props=regionprops(label_im)

print(len(props))

props[0].area #area (zero) 0th object in the image
props[0].bbox
```

```python
import matplotlib.patches as mpatches

# label image regions
label_image = label(label_im)
# to make the background transparent, pass the value of `bg_label`,
# and leave `bg_color` as `None` and `kind` as `overlay`
image_label_overlay = label2rgb(label_image, image=astronaut, bg_label=0)

fig, ax = plt.subplots(figsize=(10, 6))
ax.imshow(image_label_overlay)

for region in regionprops(label_image):
    # take regions with large enough areas
    if region.area >= 200:
        # draw rectangle around segmented object
        minr, minc, maxr, maxc = region.bbox
        rect = mpatches.Rectangle((minc, minr), maxc - minc, maxr - minr,
                                  fill=False, edgecolor='red', linewidth=2)
        ax.add_patch(rect)

ax.set_axis_off()
plt.tight_layout()
plt.show()
```
